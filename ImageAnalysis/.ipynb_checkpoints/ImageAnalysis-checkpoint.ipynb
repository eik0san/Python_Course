# Image processing with Numpy and Matplotlib

Today we will learn to do basic image processing with Python, namely with Numpy, Scipy and Matplotlib. There are also powerful libraries (Scipy-ndimage, OpenCV, Scikit-Image, PIL/Pillow...) for advanced image processing. So let's start!

First we import the libraries:

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt 
```

And load our image:

```python
im = plt.imread('lisa-simpson.jpg')
```

Let's now look at how Python sees the image.

```python
print(type(im))
print (im.shape)
print (im.ndim)
```

Python loaded the image as a 3-dimensional numpy array. This means that we can edit the image using the same techniques we already learned with numpy and matplotlib.
The first two indices give us the x and y position of the pixels, and the third gives the RGB values.

Let's display the image:

```python
plt.imshow(im)
plt.show()
```

We can crop the photo by cropping the array:

```python
im_crop = im[:225,:400,:]
plt.imshow(im_crop)
plt.show()
```

The information about the color of the image is stored in the RGB values for each pixel. Let's try to separate the channels:

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(15,5))
for c, ax in zip(range(3), ax):
    temp_im = np.zeros(im.shape, dtype="uint8") # Specifying type is useful for Matplotlib because it plots
    #differently for different data types
    temp_im[:,:,c] = im[:,:,c]
    ax.imshow(temp_im)
    #ax.set_axis_off()
```

We can think of colours as points in 3D space, so we can translate and rotate them.

```python
tempy=im[:,:,0]-5 # we subtracted 5 from the red channel accross the image
plt.imshow(tempy)
```

Converting to grayscale is a very useful operation. There are many ways to do this depeding on the features we want to enhance or supress, but the most straightforward way is taking a weighted mean of the RGB channels.

```python
# let's recall how to define a function
def to_grayscale(im, weights = np.c_[0.2989, 0.5870, 0.1140]): #the weights are normally used conversion factors
    tile = np.tile(weights, reps=(im.shape[0],im.shape[1],1)) # tile creates an array by repeating elements
    return np.sum(tile * im, axis=2)
```

```python
im_gray=to_grayscale(im)
plt.imshow(im_gray, cmap='gray')
```

We used cmap = 'gray' to plot our grayscale image. How do we know it's not cheating and we actually have a grayscale image?

```python
np.shape(im_gray)
```

Another cool operation is image thresholding. Very often when dealing with microscopy data, we want to extract and keep just some of the features in the image (for example, foreground and background). This is especially useful when dealing with fluorescence microscopy data, and it is easily done by converting the image to grayscale and applying a threshold, which gives us a binary image.

```python
def simple_threshold(im_gray, threshold=128):
    return (im_gray > threshold).astype("uint8")

thresholds = [100,120,128,150,200]

fig, ax = plt.subplots(nrows=1, ncols=len(thresholds), figsize=(20,5));
                        
for t, ax in zip(thresholds, ax):
    ax.imshow(simple_threshold(im_gray, t), cmap='gray');
    ax.set_title("Threshold: {}".format(t), fontsize=20);
    ax.set_axis_off();
```

Sometimes we want to make changes on just one segment of the image (array), and mask the rest.

```python
im_masked = np.ma.masked_where(im_gray > 120, im_gray, copy=True) # mask all indices where pixel value > 120
im_filled = np.ma.filled(im_masked, 0) # fill those pixels with 0 (black)
plt.imshow(im_filled, cmap='gray')
```

Another cool feature we can easily implement is transparency. Transparency is achieved by adding another layer of scalars (from 0 to 255) to our image, so that the RGB (x,y,3) image becomes RGBA (x,y,4). For simplicity, we will first separate the image into channels, add a transparency channel and merge them all together.

```python
red = im[:,:,0]
blue = im[:,:, 1]
green = im[:,:,2]
# now we create a transparent alpha channel and fill it with value 128 = semi-transparent
alpha= np.ones((np.shape(im)[0], np.shape(im)[1])).astype('uint8')*128
merged_image = np.dstack((red, blue, green, alpha)) # join the arrays depth-wise (2D arrays into 3D array)
# now plot it
plt.imshow(merged_image)
plt.show()
```

Ok, the image looks a bit fainter, but are we really convinced it is transparent? Let's flip the image vertically and overlay it with the original one to see the effect better.

```python
#first flip
image_flipped = np.fliplr(merged_image)
plt.imshow(image_flipped)
plt.show()
```

```python
# now overlay
plt.imshow(image_flipped)
plt.imshow(merged_image)
plt.show()
```

And this was just scratching the surface of image processing abilities of Python.
